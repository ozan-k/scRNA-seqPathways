```python
import json
import os
import plotly.express as px
import numpy as np
import math

def get_data(file_name):
    with open(file_name,encoding="utf-8") as json_file:
        data = json.load(json_file)
    return data

def add_all(this_dct):
    lst = []
    for d in this_dct:
        lst += this_dct[d]
    return lst
```

```python
data = get_data("./distributions.json")
dct = {}
for d in data["singles"]:
    dct[d] = add_all(data["singles"][d])

for d in dct:        
    print(d)

l1 = [ math.log(k) for k in dct["PathwaysOnlyHealthy"]]
l2 = [ math.log(k) for k in dct["PathwaysOnlyInflamed"]]
l3 = [ math.log(k) for k in dct["PathwaysOnlyNoninflamed"]]
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

df =pd.DataFrame(dict(
    series=np.concatenate((["Healthy"]*len(l1), ["Inflamed"]*len(l2),["Non-inflamed"]*len(l3))), 
    data  =np.concatenate((l1,l2,l3))
))

fig = px.histogram(df, x="data", color="series", barmode="overlay")
fig.update_xaxes(title_text='Score (log)')
fig.update_layout(title="Expressed pathways")
```

```python
for d in data["doubles"]:
    dct[d] = add_all(data["doubles"][d])

for d in dct:        
    print(d)

l4 = [ math.log(k) for k in dct["Pathways__Healthy_and_Inflamed"]]
l5 = [ math.log(k) for k in dct["Pathways__Healthy_and_Noninflamed"]]
l6 = [ math.log(k) for k in dct["Pathways__Inflamed_and_Noninflamed"]]
    
df =pd.DataFrame(dict(
    series=np.concatenate((["Healthy & Inflamed"]*len(l4), ["Healthy & Non-inflamed"]*len(l5),["Inflamed & Non-inflamed"]*len(l6))), 
    data  =np.concatenate((l4,l5,l6))
))

fig = px.histogram(df, x="data", color="series", barmode="overlay")
fig.update_xaxes(title_text='Score (log)')
fig.update_layout(title="Expressed pathways")
# fig.title("saas") 
```

```python
for d in data["all"]:
    dct[d] = add_all(data["all"][d])

for d in dct:        
    print(d)

l7 = [ math.log(k) for k in dct["PathwaysInAllConditions"]]

fig = px.histogram(x = np.array(l7))
fig.update_xaxes(title_text='Score (log)')
fig.update_layout(title="Expressed pathways for Healthy & Inflamed & Non-inflamed")
fig.show()
```
